```python
from matplotlib import pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
import itertools
```

```python
plt.figure(figsize=(8, 5))

x = np.arange(0,10,0.1)
y = np.arange(-1,1,0.1)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)

#第３引数で指定した高さとグラフに囲まれた領域が塗りつぶされる
plt.fill_between(x,y1,y2,facecolor='r', alpha=0.5)
plt.fill_between(x,y1,-1,facecolor='g', alpha=0.2)
plt.fill_between(x,y1,1,facecolor='b', alpha=0.2)

plt.grid(True)
plt.show()
```

```python
iris = datasets.load_iris()

# 特徴量の入ったデータ (4 次元)
features = iris.data
# 各特徴量の名前
feature_names = iris.feature_names
# データと品種の対応
targets = iris.target

# グラフの全体サイズを指定する
plt.figure(figsize=(12, 12))
# ラベルが3種類で、各ラベルに合わせてマークと色を指定する
for t, marker, c in zip(range(3), '>ox', 'rgb'):
    plt.scatter(
        features[targets == t, 2],
        features[targets == t, 3],
        marker=marker,
        c=c,
    )

plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.autoscale()
plt.grid(True)
    
plt.show()
```

```python
# 特徴量の入ったデータ (4 次元)
features = iris.data
# 各特徴量の名前
feature_names = iris.feature_names
# データと品種の対応
targets = iris.target

# グラフの全体サイズを指定する
plt.figure(figsize=(12, 12))
# ラベルが3種類で、各ラベルに合わせてマークと色を指定する
for t, marker, c in zip(range(3), '>ox', 'rgb'):
    plt.scatter(
        features[targets == t, 2],
        features[targets == t, 3],
        marker=marker,
        c=c,
    )

plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.grid(True)

iris = datasets.load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target
df_iris['target_name'] = iris.target_names[iris.target]
df_iris.columns = ['sepal_l', 'sepal_w', 'petal_l', 'petal_w','target','target_name']

#平均値
x_mean = df_iris['petal_l'].mean()
y_mean = df_iris['petal_w'].mean()
# 最大値、最小値
x_max = df_iris['petal_l'].max()
x_min = df_iris['petal_l'].min()
y_max = df_iris['petal_w'].max()
y_min = df_iris['petal_w'].min()
# やり方正しいかわからない
# 平均値からの絶対値の差分を20%増しで範囲を取る
x_max_size = x_mean + (abs(x_max) - abs(x_mean))*1.2
x_min_size = x_mean - (abs(x_mean) - abs(x_min))*1.2
y_max_size = y_mean + (abs(y_max) - abs(y_mean))*1.2
y_min_size = y_mean - (abs(y_mean) - abs(y_min))*1.2

# グラフ描画範囲
x = np.arange(x_min_size,x_max_size,0.1)
y = np.arange(y_min_size,y_max_size,0.1)

y1 = np.sin(x) + y_mean
y2 = np.cos(x) + y_mean

plt.plot(x, y1)
plt.plot(x, y2)

#第３引数で指定した高さとグラフに囲まれた領域が塗りつぶされる
plt.fill_between(x,y1,y2,facecolor='r', alpha=0.5)
plt.fill_between(x,y1,y_min_size,facecolor='g', alpha=0.2)
plt.fill_between(x,y1,y_max_size,facecolor='b', alpha=0.2)
plt.show()
```
