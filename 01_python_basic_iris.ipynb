```python
# 特徴量の入ったデータ (4 次元)
features = iris.data
# 各特徴量の名前
feature_names = iris.feature_names
# データと品種の対応
targets = iris.target

# グラフの全体サイズを指定する
plt.figure(figsize=(12, 8))

# ラベルの組み合わせ生成[4C2]
for i,(x,y) in enumerate(itertools.combinations(range(4), 2)):
    
    # サブグラフ
    plt.subplot(2, 3, i+1)
    
    # irisデータ全件に対して、ラベルの確認と色とマーク付け
    for i in range(len(features)):
        if(targets[i] == 0):
            marker = '>'
            c = 'r'
        elif(targets[i] == 1):
            marker = 'o'
            c = 'g'
        elif(targets[i] == 2):
            marker = 'x'
            c = 'b'
        plt.scatter(
            features[i,x],
            features[i,y],
            marker=marker,
            c=c,
        )

    plt.xlabel(feature_names[x])
    plt.ylabel(feature_names[y])
    plt.autoscale()
    plt.grid(True)
    
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.title("target_map")

# ラベルが3種類で、各ラベルに合わせてマークと色を指定する
# featuresの中から[targets == t] で一致したものだけのリストが作られ、
# range(3)なので、0,1,2の3つのリストが作られる
for t, marker, c in zip(range(3), '>ox', 'rgb'):
    plt.scatter(
        features[targets == t, x],
        features[targets == t, y],
        marker=marker,
        c=c,
    )

plt.xlabel(feature_names[2])
plt.ylabel(feature_names[3])
plt.autoscale()
plt.grid(True)
    
plt.show()
```

```python
# 特徴量の入ったデータ (4 次元)
features = iris.data
# 各特徴量の名前
feature_names = iris.feature_names
# データと品種の対応
targets = iris.target

# グラフの全体サイズを指定する
plt.figure(figsize=(12, 8))

# 二次元のグラフを作りたいので特徴量の組み合わせを作る
for i, (x, y) in enumerate(itertools.combinations(range(4), 2)):
    # サブグラフ
    plt.subplot(2, 3, i + 1)
    # 各品種はマーカーの色や形を変える
    for t, marker, c in zip(range(3), '>ox', 'rgb'):
        plt.scatter(
            features[targets == t, x],
            features[targets == t, y],
            marker=marker,
            c=c,
        )
        plt.xlabel(feature_names[x])
        plt.ylabel(feature_names[y])
        plt.autoscale()
        plt.grid()

plt.show()
```

# pandas data_frame操作

```python
#iris.data          # 特徴量の入ったデータ (4 次元) 
#iris.feature_names # 各特徴量の名前
#iris.target        # データと品種の対応
#iris.target_names  # 品種の名前（文字列）

iris = datasets.load_iris()
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target
df_iris['target_name'] = iris.target_names[iris.target]
df_iris.head()
```

```python
# 行名、列名の変更
new_df_iris=df_iris.rename(columns={'sepal length (cm)' : 'sepal_l'}, index={0:'zero'})
new_df_iris.head()
```

```python
#列名の複数変更、dataframeの更新
new_df_iris.rename(columns={'sepal length (cm)' : 'sepal_l',
                            'sepal width (cm)'  : 'sepal_w',
                            'petal length (cm)' : 'petal_l',
                            'petal width (cm)'  : 'petal_w'
                           }, inplace=True)
new_df_iris.head()
```

```python
# 列名の一括変更
df_iris.columns = ['sepal_l', 'sepal_w', 'petal_l', 'petal_w','target','target_name']
df_iris.head()
```

```python
print(len(df_iris))  #行数の確認
print(df_iris.shape) #次元数の確認（行数,列数）

# 各列の基礎統計量の確認
# Rでいうところのsummary()
print(df_iris.describe()) # 平均、分散、4分位など

df_iris.plot()
df_iris.plot(kind='box')
df_iris.plot.scatter(x='sepal_l', y='petal_w', alpha=0.5)
```

```python
df_iris.plot.scatter(x='sepal_l', y='petal_w',
                     c=df_iris['target'],
                     colormap='Set1',
                     legend=True,
                     alpha=1.0)
plt.grid()
plt.show()
```
