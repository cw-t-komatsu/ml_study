```python
import matplotlib.pyplot as plt
from sklearn import cluster
from sklearn import datasets
```

```python
# iris データセットをロード
iris = datasets.load_iris()
X = iris.data
```

```python
# k-means モデルの作成
# クラスタ数は 3 を指定
model = cluster.KMeans(n_clusters=3)
model.fit(X)
# クラスタリング結果ラベルの取得
labels = model.labels_
```

```python
# 1 番目のキャンバスを作成
plt.figure(1)

# ラベル 0 の描画
ldata = X[labels == 0]
plt.scatter(ldata[:, 2], ldata[:, 3], color='green')
# ラベル 1 の描画
ldata = X[labels == 1]
plt.scatter(ldata[:, 2], ldata[:, 3], color='red')
# ラベル 2 の描画
ldata = X[labels == 2]
plt.scatter(ldata[:, 2], ldata[:, 3], color='blue')
# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.show()
```

```python
# ラベル 0 の描画
ldata = X[iris.target == 0]
plt.scatter(ldata[:, 2], ldata[:, 3], color='red')

# ラベル 1 の描画
ldata = X[iris.target == 1]
plt.scatter(ldata[:, 2], ldata[:, 3], color='green')

# ラベル 2 の描画
ldata = X[iris.target == 2]
plt.scatter(ldata[:, 2], ldata[:, 3], color='blue')

# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

plt.show()
```

```python
plt.figure(figsize=(14, 4))

# K-meansで分類した結果の表示
plt.subplot(1, 2, 1)
# ラベル 0 の描画
ldata = X[labels == 0]
plt.scatter(ldata[:, 2], ldata[:, 3], color='green')
# ラベル 1 の描画
ldata = X[labels == 1]
plt.scatter(ldata[:, 2], ldata[:, 3], color='red')
# ラベル 2 の描画
ldata = X[labels == 2]
plt.scatter(ldata[:, 2], ldata[:, 3], color='blue')
# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


# 正解データの表示
plt.subplot(1, 2, 2)
# ラベル 0 の描画
ldata = X[iris.target == 0]
plt.scatter(ldata[:, 2], ldata[:, 3], color='red')
# ラベル 1 の描画
ldata = X[iris.target == 1]
plt.scatter(ldata[:, 2], ldata[:, 3], color='green')
# ラベル 2 の描画
ldata = X[iris.target == 2]
plt.scatter(ldata[:, 2], ldata[:, 3], color='blue')
# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

plt.show()
```

# irisの次元数を４から２に減らして実験

```python
# iris データセットをロード
iris = datasets.load_iris()
X = iris.data[:,2:4]

# k-means モデルの作成
# クラスタ数は 3 を指定
model = cluster.KMeans(n_clusters=3)
model.fit(X)
# クラスタリング結果ラベルの取得
labels = model.labels_
```

```python
plt.figure(figsize=(14, 4))

# K-meansで分類した結果の表示
plt.subplot(1, 2, 1)
# ラベル 0 の描画
ldata = X[labels == 0]
plt.scatter(ldata[:, 0], ldata[:, 1], color='green')
# ラベル 1 の描画
ldata = X[labels == 1]
plt.scatter(ldata[:, 0], ldata[:, 1], color='red')
# ラベル 2 の描画
ldata = X[labels == 2]
plt.scatter(ldata[:, 0], ldata[:, 1], color='blue')
# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])


# 正解データの表示
plt.subplot(1, 2, 2)
# ラベル 0 の描画
ldata = X[iris.target == 0]
plt.scatter(ldata[:, 0], ldata[:, 1], color='green')
# ラベル 1 の描画
ldata = X[iris.target == 1]
plt.scatter(ldata[:, 0], ldata[:, 1], color='blue')
# ラベル 2 の描画
ldata = X[iris.target == 2]
plt.scatter(ldata[:, 0], ldata[:, 1], color='red')
# x軸、y軸の設定
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])

plt.show()
```
